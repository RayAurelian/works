```python
import pandas as pd ; import matplotlib.pyplot as plt ; import numpy as npy ; from io import BytesIO
import requests 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PowerTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import RobustScaler
from sklearn import metrics
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
import seaborn as sns
```

```python
df=pd.read_csv('Laptop_price.csv')
price=df[['Price']]
df
model= df.drop('Price', axis = 1) 
```

```python

X_train, X_test,Y_train, Y_test = train_test_split(model,price, test_size=0.50, random_state=npy.random.randint(1000000)) 
```

```python
num_colomns=list(X_train.select_dtypes(exclude=['object']).columns)
cat_columns=list(X_train.select_dtypes(include=['object']).columns)
numerical = Pipeline(steps=[
    ("simple_imputer", SimpleImputer(strategy='mean')),
    ("power_transform", PowerTransformer()),
    ("Scaler", StandardScaler())
])
categorical = Pipeline(steps=[
    ("simple_imputer", SimpleImputer(strategy='most_frequent')),
    ('OneHotEncoder', OneHotEncoder(handle_unknown='ignore',
                                    sparse_output=False))
])
ct=ColumnTransformer([
    ('numerical',numerical,num_colomns),
    ("categorical", categorical, cat_columns)
])
pipe=Pipeline(steps=[
    ("ct",ct),
    ("XGBRegressor", XGBRegressor())
])
param_grid={
    'XGBRegressor__n_estimators': [10, 11 , 20 ,  50], 
    'XGBRegressor__learning_rate': [0.1 , 0.111 , 0.333 , 0.666],
    'XGBRegressor__max_depth': [2, 5 , 6 , 15],
    'XGBRegressor__gamma': [0.1 , 0.111 , 0.333 , 0.666 ], 
    'XGBRegressor__booster': ['gbtree', 'gblinear'],
    'ct__numerical': [StandardScaler(), RobustScaler()],
}
```

```python
clf = GridSearchCV(pipe,param_grid,verbose=3) 
```

```python
clf.fit(X_train,Y_train)
```

```python
clf.score(X_test,Y_test)
```

```python
modelstat=clf.best_estimator_
modelstat
```
